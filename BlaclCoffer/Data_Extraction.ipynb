### <u>Now loop through multiple url </u>

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import os

def extract_article_info(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, 'html.parser')

        # Try to find heading using different possible tags/classes
        heading = soup.find('h1')
        if not heading:
            heading = soup.find('h2')
        if not heading:
            heading = soup.title
        heading_text = heading.get_text().strip() if heading else "Heading not found"

        # Try to find article body using different possible div classes
        article_body = soup.find('div', class_='td-post-content')
        if not article_body:
            article_body = soup.find('div', class_='entry-content')
        if not article_body:
            article_body = soup.find('article')
        paragraphs = article_body.find_all('p') if article_body else []
        article_text = ' '.join([para.get_text().strip() for para in paragraphs]) if paragraphs else "Article text not found"

        return heading_text, article_text

    except requests.exceptions.RequestException as e:
        print(f"Failed to fetch {url}: {e}")
        return "Error", "Error"

```

```python
def save_article_to_file(url_id, url, heading, text, output_dir='articles'):
    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)
    # Create a valid filename using the URL ID
    filename = f"{url_id}_output.txt"
    filepath = os.path.join(output_dir, filename)

    with open(filepath, 'w', encoding='utf-8') as file:
        file.write(f"Heading: {heading}\n\n")
        file.write(f"Article Text:\n{text}\n")


```

```python
for index, row in input_df.iterrows():
    url_id = f"blackassign{index+1:04d}"  # Creating a unique ID based on the row index
    url = row['URL']  # Assuming the URLs are in a column named 'URL'
    heading, text = extract_article_info(url)
    save_article_to_file(url_id, url, heading, text)
    print(f"Processed URL_ID {url_id}. Data saved to {url_id}_output.txt")

print("All articles have been extracted and saved.")
```

```python
import os

def check_extracted_files(output_dir=r"C:\Users\dell\OneDrive\Sem5\artilce\articles"):
    empty_files = []

    # List all files in the output directory
    for filename in os.listdir(output_dir):
        if filename.endswith('_output.txt'):
            filepath = os.path.join(output_dir, filename)
            # Check if the file is empty
            if os.path.getsize(filepath) == 0:
                empty_files.append(filename)
            else:
                with open(filepath, 'r', encoding='utf-8') as file:
                    content = file.read().strip()
                    if content.startswith("Article Text:\n") and len(content.splitlines()) <= 2:
                        empty_files.append(filename)  # Likely empty content

    if empty_files:
        print("The following files are empty or have no article text:")
        for file in empty_files:
            print(file)
    else:
        print("All files contain article text.")

check_extracted_files()
```
